```python
pickup = df[['id','pickup_datetime','pickup_longitude','pickup_latitude','trip_duration']]
dropoff = df[['id','dropoff_datetime','dropoff_longitude','dropoff_latitude','trip_duration']]
```

```python
# pickup locations 
```

```python
dataframe = pickup

hmap = folium.Map(location=[40.7127837,-74.0059413], zoom_start=15)
heat_data = [[row['pickup_latitude'],row['pickup_longitude']] for index, row in pickup[:1000].iterrows()]
HeatMap(heat_data).add_to(hmap)

hmap
```
